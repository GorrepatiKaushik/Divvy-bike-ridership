```python
petrol_initial = pd.read_csv("Data/petrol price.csv")
```

```python
petrol_initial["Date"] = pd.to_datetime(petrol_initial["Date"])

petrol_initial["Month Year"] =  petrol_initial['Date'].dt.to_period('M')

petrol_initial['Month Year'] = petrol_initial["Month Year"].dt.strftime('%B %Y')
```

```python
petrol_initial.columns = ["Date", "Price", "Month Year"]

```

```python
petrol_initial["Month Year"] = pd.Categorical(petrol_initial["Month Year"], months_list)
```

```python
petrol_initial = petrol_initial.groupby(["Month Year"])["Price"].mean().round(3).reset_index(name = "Monthly avg")
```

```python
temp13_19_final = temp13_19[["usertype","Month Year","number of rides"]]
temp13_19_final = temp13_19.pivot(columns = "usertype", index="Month Year", values="number of rides").reset_index()
temp13_19_final["Montly avg"] = petrol_initial["Monthly avg"]
#print(temp13_19_final)
```

```python
import matplotlib.patches as mpatches

orange_patch = mpatches.Patch(color='orange', label='Member rides')
green_patch = mpatches.Patch(color='lightgreen', label='Casual rides')
fig,ax = plt.subplots(figsize=(100,30))

x = np.arange(len(temp13_19_final))
ax.bar(x-0.2, temp13_19_final["casual"], width, color="lightgreen", edgecolor="green")
ax.bar(x+0.2, temp13_19_final["member"], width, color="orange", edgecolor="orange")
ax.set_xticklabels(temp13_19_final["Month Year"],  rotation=80, fontsize=40)

# set x-axis label
ax.set_xlabel("Month-Year",fontsize=90)
ax.tick_params(axis='y', labelsize=50 )

# set y-axis label
ax.set_ylabel("Number of rides", fontsize=90)

ax.grid(axis="y", alpha=0.3)
ax.set_title("Monthly member rides vs casual rides from 2013 through 2019 with respect to petrol price", fontsize=110)
ax.legend(handles=[orange_patch, green_patch], fontsize=50)

ax2=ax.twinx()

# make a plot with different y-axis using second axis object
ax2.plot(petrol_initial["Month Year"], list(petrol_initial["Monthly avg"]),color="blue",marker="o", linewidth=5)
ax2.set_ylabel("Petrol price (USD per gallon)",color="blue", fontsize=90)

ax2.tick_params(axis='y', labelsize=50 )
# save the plot as a file
fig.savefig('Monthly.jpg',
            format='jpeg',
            dpi=100,
            bbox_inches='tight')
```
